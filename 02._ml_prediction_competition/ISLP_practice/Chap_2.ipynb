```python
# Three-dimensional data representation- Contour plot

fig, ax = subplots(figsize=(8, 8))
x = np.linspace(-np.pi, np.pi, 50)   # NumPy divides the interval from -π to π into 50 evenly spaced values. this creates grid resolution
y = x
f = np.multiply.outer(np.cos(y), 1 / (1 + x**2))   # Compute a 2D grid (50x50) of function values that represent a surface. note both are vectors of length 50
# 1 / (1 + x**2) Create values that get smaller as x moves away from 0.
# If you use outer multiplication, every element of the first array multiplies every element of the second array.
# outer multiplication is allowed when both inputs are 1 dimensional arrays


# Draw a contour plot x → horizontal coordinates,y → vertical coordinates,f → function values
ax.contour(x, y, f);                       
# you are evaluating the function over many combinations of x-values and y-values 
# 
# Inside each grid square it does linear interpolation. variant of marching squares algorith       
```

```python
# increase contour resolution of image by adding more levels

fig, ax = subplots(figsize=(8, 8))
ax.contour(x, y, f, levels=45);  #more levels

# there are two different kinds of resolution involved in contour plots:
# grid resolution (how finely the function is sampled)
# contour resolution (how many contour lines are drawn)

# note Contour plots do not show height directly. Instead they show curves where the height is the same.
```

```python
######################### Lets study Indexing Data ################

# Create two dimensional array
A = np.array(np.arange(16)).reshape((4, 4))
print(A)

A[1,2] 
```

```python
# Select multiple rows

print(A[[1,3]]) # we select row 1 and 3

# Select multiple columns

print(A[:,[0,2]]) # we select column 0 and 2


```

```python
# Submatrix indexing  

idx = np.ix_([1,3],[0,2,3])   # rows 2nd and 4th, columns 1st and 3rd and 4th
A[idx]

# alternatively use A[[1,3]][:,[0,2]] , it gets rows first and then slices for columns

# another alternative is A[1:4:2,0:3:2] where the syntax is "start : stop : step" for both row slice and column slice

```

```python
######################### Lets study Boolean Indexing  #########################

keep_rows = np.zeros(A.shape[0], bool)  # array of size A.shape[0] ie 4 rows from previous case thus we get array of size 4
print(keep_rows)                               # np.zeros creates an araay of zeros

keep_rows[[1,3]] = True                 # change elements 2 and 4 to true
print(keep_rows)
```

```python
# Boolean equality check

np.all(keep_rows == np.array([0,1,0,1]))   # the `==`   operation is applied elementwise.

```

```python
# Test np.any()

print(np.any(keep_rows))                
# True if ANY element in keep_rows is True

print(np.any(keep_rows == np.array([0,1,0,1])))   
# compare elementwise with [0,1,0,1], then check if ANY comparison is True

print(np.any(keep_rows == np.array([1,0,1,0])))   
# compare elementwise with [1,0,1,0], then check if ANY comparison is True
```

```python
# even though `np.array([0,1,0,1])`  and `keep_rows` are equal according to `==`, they index different sets of rows!

print(A[np.array([0,1,0,1])]) # row 0 and 1 and 0 and 1

A[keep_rows]  # row 1 and 3 ie only true values returned
```

```python
# Submatrix work with Booleans 

keep_cols = np.zeros(A.shape[1], bool)
keep_cols[[0, 2, 3]] = True

keep_rows = np.zeros(A.shape[0], bool)                              
keep_rows[[1,3]] = True   

idx_bool = np.ix_(keep_rows, keep_cols)
A[idx_bool]

# (kinda complicated but keep for reference, generally use .ix_ directly)
```

```python
##############################  Let's study For loops ############################## 

total = 0
for value in [3,2,19]:
    total += value
print('Total is: {0}'.format(total))
```

```python
# nested loop

total = 0
for value in [2,3,19]:
    for weight in [3, 2, 1]:
        total += value * weight
print(f"Total is: {total}")
```

```python
# Weighted Sum Shortcut using ZIP function

total = 0
for value, weight in zip([2,3,19],
                         [0.2,0.3,0.5]):
    total += weight * value

print(f"Weighted average is: {total}")
```

```python
##############################  Let's study String Formatting ############################## 

# Calculate Percentage of NANs in each column

rng = np.random.default_rng(1)
A = rng.standard_normal((127, 5))
M = rng.choice([0, np.nan], p=[0.8,0.2], size=A.shape)
A += M
D = pd.DataFrame(A, columns=['food',
                             'bar',
                             'pickle',
                             'snack',
                             'popcorn'])
print( D[:3] )

# Below is the loop tha acheves it

for col in D.columns:
    print(f'Column "{col}" has {D[col].isna().mean():.2%} missing values')    

# Inside an f-string, anything between {} can have two parts: {expression:format_spec}
# expression   = D[col].isna().mean() # produces a number, usually a decimal between 0 and 1.
# format_spec  = .2% # % means percentage form, .2 means exactly 2 decimal places

# .isna() Creates a boolean Series where: True = value is NaN , False = value is not missing.
# .mean() In pandas, booleans behave like numbers: True = 1, False = 0 So the mean of the boolean series equals the fraction of missing values
```
