```python
# Graphical Diagnostics of Linear Regression

options = list(range(1,7))

linRegGraphDict = {
    1:"1. Nonlinearity of response predictor relationships",
    2:"2. Correlation of error terms",
    3:"3. Nonconstant variance of error terms",
    4:"4. Outliers",
    5:"5. High-leverage points",
    6:"6. Collinearity"
}

# dictionary just for plots names
plotDict = {
    1: ["Residuals vs Fitted", "Partial residual plots"],
    2: ["Residuals vs Time"],
    3: ["Residuals vs Fitted", "Scale-Location plot"],
    4: ["Residuals vs Fitted", "QQ plot"],
    5: ["Residuals vs Leverage", "Cook's Distance"],
    6: ["Scatterplot matrix", "VIF", "Partial regression plots"]
}

choice = input(
    "Potential problems with linear regression\n\
        Choose an option:\n\n (simply type digit like '2' for )\n\n"
    "1. Nonlinearity of response predictor relationships\n"
    "2. Correlation of error terms\n"
    "3. Nonconstant variance of error terms\n"
    "4. Outliers\n"
    "5. High-leverage points\n"
    "6. Collinearity\n"
)

if choice.isdigit() and int(choice) in options:
    choice = int(choice)

    print("You selected:", linRegGraphDict[choice])

    print("\nRelevant diagnostic plots:")
    for plot in plotDict[choice]:
        print("-", plot)

    if choice == 3:
        print("\n Graphical Diagnostics: Does the residuals vs. fitted plot show a constant variance or non-constant?")

        Res_Time_plt_results = input("\n Type n for non-constant, or anything else for constant")

        if Res_Time_plt_results == "n":

            if USE_LATEX:
                print("\n You have chosen non-constant variance, thus variance of each raw observations is sig_i ^2_i = sig^2 / n_i \n\nIf you would you like to remedy that, use weighted least square with weights proportional to the inverse variances \n- i.e. w_i = n_i,\nMost linear regression software allows for observation weights ")

                from IPython.display import display, Math
                display(Math(r"\sigma_i^2 = \frac{\sigma^2}{n_i}"))
                display(Math(r"w_i = n_i"))

            else:
                print("\n You have chosen non-constant variance, thus variance of each raw observations is sig_i ^2_i = sig^2 / n_i \n\nIf you would you like to remedy that, use weighted least square with weights proportional to the inverse variances \n- i.e. w_i = n_i,\
                   \nMost linear regression software allows for observation weights ")
                
    if choice == 4:
        # Anomaly detection via studentized residuals (plot)
        print("\nIf difficult to decide how large a residual to consider an outlier, then one useful rule is to look at studentized residuals.\n\
        Observations with studentized residuals above 3 in absolute value are often flagged as potential outliers.\n")

        student_resid_check = input("Do you notice any studentized residuals crossing 3 in absolute value? type [y/n]\n")

        if student_resid_check == "y":
            print("\nYou may have potential outliers.\n\
                  \nA common decision rule is that observations with |studentized residual| > 3 deserve further investigation.\
                  \nThese points may indicate unusual response values relative to what the model predicts.\n")

        elif student_resid_check == "n":
            print("\nNo obvious outliers are indicated by the studentized residual > 3 rule.\n")

        else:
            print("\nInvalid input\n")
        

else:
    print("Invalid input")
```
