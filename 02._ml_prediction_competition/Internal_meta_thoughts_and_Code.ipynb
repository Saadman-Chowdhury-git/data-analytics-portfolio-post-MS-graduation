## Classifier Model Selection Heuristics

- Small data → Naive Bayes, LDA, Logistic  
- High-dimensional (p ≫ n) → Naive Bayes, Regularized Logistic  
- Nonlinear boundary → QDA, k-NN, Trees, Boosting  
- Correlated features → Logistic, LDA, Trees  
- Interpretability → Logistic, LDA

```python
# Simulate data and compare classifiers with boxplots of test error rate

%pip install scikit-learn

import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier, GradientBoostingClassifier
from sklearn.metrics import accuracy_score

# Define classifiers
models = {
    "Naive Bayes": GaussianNB(),
    "LDA": LinearDiscriminantAnalysis(),
    "QDA": QuadraticDiscriminantAnalysis(reg_param=0.1),  # regularized to avoid singularity
    "Binary Logistic": LogisticRegression(max_iter=1000),
    "k-NN": KNeighborsClassifier(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "Gradient Boosting": GradientBoostingClassifier()
}

# Store results
results = {name: [] for name in models}

# Simulation loop
n_simulations = 50

for _ in range(n_simulations):
    X, y = make_classification(
        n_samples=500,
        n_features=10,
        n_informative=5,
        n_redundant=2,
        n_classes=2
    )
    
    X_train, X_test, y_train, y_test = train_test_split(
        X, y, test_size=0.3
    )
    
    for name, model in models.items():
        model.fit(X_train, y_train)
        y_pred = model.predict(X_test)
        error = 1 - accuracy_score(y_test, y_pred)
        results[name].append(error)

# Prepare data for boxplot
data = [results[name] for name in models]

# Plot
plt.figure()
plt.boxplot(data, labels=list(models.keys()))
plt.xticks(rotation=45)
plt.ylabel("Test Error Rate")
plt.title("Classifier Comparison (Test Error Rates)")
plt.show()
```

# Linear Model Selection Notes

- **OLS** is the clean classical baseline when the feature space is small and stable.
- **Ridge** is usually preferred when predictors are numerous and correlated.
- **Lasso** is useful when you believe only a subset of predictors should remain in the model.
- **Elastic Net** is often the safest practical choice when you want both regularization and selection in the presence of correlated predictors.

# Practical Rule of Thumb

- Start with **OLS** if the problem is small and interpretable.
- Move to **Ridge** if coefficients become unstable or predictors are collinear.
- Move to **Lasso** if simplification and feature selection matter.
- Choose **Elastic Net** if predictors are correlated and you still want sparsity.

# Linear Model Selection Heuristics

- Simple linear relationship, strongest interpretability → OLS
- Many correlated predictors → Ridge
- Need feature selection / sparsity → Lasso
- Want both shrinkage and variable selection → Elastic Net
- High-dimensional setting (p ≫ n) → Ridge or Elastic Net
- Strong multicollinearity → Ridge or Elastic Net
- Small sample with many weak predictors → Ridge
- Small sample with only a few true signals expected → Lasso or Elastic Net

```python
# Validation Set Approach (Repeated): Polynomial Degree Selection


import numpy as np
import matplotlib.pyplot as plt

np.random.seed(42)

n = 70
x = np.random.uniform(-4, 4, n)
y = 6 + 2.5*x - 2.8*x**2 + np.random.normal(0, 6, n)

degrees = range(1, 11)
all_mse = []

for i in range(10):
    idx = np.random.permutation(n)
    split = n // 2

    train_idx = idx[:split]
    val_idx = idx[split:]

    mse_row = []

    for d in degrees:
        X_train = np.column_stack([x[train_idx]**k for k in range(d + 1)])
        X_val = np.column_stack([x[val_idx]**k for k in range(d + 1)])

        beta = np.linalg.pinv(X_train) @ y[train_idx]
        y_pred = X_val @ beta

        mse = np.mean((y[val_idx] - y_pred)**2)
        mse_row.append(mse)

    all_mse.append(mse_row)

all_mse = np.array(all_mse)

mean_mse = all_mse.mean(axis=0)
std_mse = all_mse.std(axis=0)

print("Degree | Mean MSE | Std MSE")

for d in degrees:
    print(f"{d:>6} | {mean_mse[d-1]:>8.2f} | {std_mse[d-1]:>7.2f}")

plt.figure(figsize=(8, 5))

for row in all_mse:
    plt.plot(degrees, row, alpha=0.5)

plt.plot(degrees, mean_mse, linewidth=2)

upper = np.percentile(all_mse, 90)
plt.ylim(0, upper)

plt.xlabel("Polynomial Degree")
plt.ylabel("Validation MSE")
plt.title("Validation MSE Across 10 Validation Splits")
plt.xticks(list(degrees))
plt.show()
```
