```python
# ──────────────────────────────────────────── Model Fitting ──────────────────────────────────────────

'''
Context: compute OLS estimates using the normal equation; foundation for all later regression extensions
Plain-English: construct design matrix and solve for beta using linear algebra
Real-world: identical structure used in generalized linear models and many ML pipelines
Reappears: yes — this formulation extends directly to multiple regression
'''

import numpy as np  # core linear algebra operations

# original data
X_raw = np.array([1, 2, 3, 4, 5])  # predictor
Y = np.array([2, 4, 5, 4, 5])      # response

# construct design matrix with intercept term
X = np.column_stack((np.ones(len(X_raw)), X_raw))  # adds bias column

# compute (X^T X)
XtX = X.T @ X  # Gram matrix

# compute (X^T Y)
XtY = X.T @ Y  # cross-product with response

# solve normal equation
beta_hat = np.linalg.inv(XtX) @ XtY  # OLS estimator

'''
Interpretation: beta_hat is a length-2 vector [β0_hat, β1_hat]; failure here typically indicates singular XtX (perfect collinearity or insufficient variation in X); numerical instability may arise for poorly scaled data, motivating QR/SVD in later extensions
'''

print()
```
